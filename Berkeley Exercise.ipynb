```python
import pandas as pd
import numpy as np
df=pd.read_csv("Berkeley.csv")
df
```

```python
genderTotal=df.groupby("Gender").sum()
genderTotal
```

```python
maleData=df[df.Gender=="Male"]
maleData
```

```python
femaleData=df[df.Gender=="Female"]
femaleData
```

```python
female_admitted,female_rejected=femaleData.loc[femaleData["Admit"]=="Admitted"],femaleData.loc[femaleData["Admit"]!="Admitted"]
female_admitted
female_rejected
```

```python
male_admitted,male_rejected=maleData.loc[maleData["Admit"]=="Admitted"],maleData.loc[maleData["Admit"]!="Admitted"]
male_admitted
male_rejected
```

```python
percent_male_admitted='{0:.2%}'.format(np.float(male_admitted.Freq.sum()/maleData.Freq.sum()))
percent_male_rejected='{0:.2%}'.format(np.float(male_rejected.Freq.sum()/maleData.Freq.sum()))
percent_female_admitted='{0:.2%}'.format(np.float(female_admitted.Freq.sum()/femaleData.Freq.sum()))
percent_female_rejeted='{0:.2%}'.format(np.float(female_rejected.Freq.sum()/femaleData.Freq.sum()))
```

```python
summary=[{"Gender":"Male","Total":maleData.Freq.sum(),"Admitted":male_admitted.Freq.sum(),"Rejected":male_rejected.Freq.sum(),"Percentage Admitted":percent_male_admitted,"Percentage Rejected":percent_male_rejected},
        {"Gender":"Female","Total":femaleData.Freq.sum(),"Admitted":female_admitted.Freq.sum(),"Rejected":female_rejected.Freq.sum(),"Percentage Admitted":percent_female_admitted,"Percentage Rejected":percent_female_rejeted}]
```

```python
myDf=pd.DataFrame(summary)
myDf
```
